```python
import pandas as pd  # pandas
import matplotlib.pyplot as plt  # matplotlibpyplot

# Excel
df = pd.read_excel('rice.xlsx')
# 
plt.rcParams['font.sans-serif'] = ['SimHei']
# 
plt.figure(figsize=(7, 5))
labels = df['']  # 
per = df['']  # 
# 
colors = ['red', 'yellow', 'green', 'blue', 'plum', 'lavender',
          'pink', 'cyan', 'teal', 'orange', 'coral']
# 
plt.pie(per, labels=labels, colors=colors,  # 
        autopct='%1.1f%%', startangle=90,  # y
        radius=1.1, center=(0.2, 0.2),  # 1.1(0.2, 0.2)
        textprops={'fontsize': 10, 'color': 'black'},  # 10
        pctdistance=0.6)  # 
plt.title('202011', size=18)  # 18
plt.show()  # 
```
